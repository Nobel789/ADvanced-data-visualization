```python
plt.hist(Mo[Mo['Genre'] == 'Action']['Budget in million'], label='Action', alpha=0.5)
plt.hist(Mo[Mo['Genre'] == 'Comedy']['Budget in million'], label='Comedy', alpha=0.5)
plt.hist(Mo[Mo['Genre'] == 'Adventure']['Budget in million'], label='Adventure', alpha=0.5)
plt.hist(Mo[Mo['Genre'] == 'Romance']['Budget in million'], label='Romance', alpha=0.5)

# Adding legend
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
#with bins
plt.hist(Mo[Mo['Genre'] == 'Action']['Budget in million'], label='Action',bins=15, alpha=0.5)
plt.hist(Mo[Mo['Genre'] == 'Comedy']['Budget in million'], label='Comedy', bins=15,alpha=0.5)
plt.hist(Mo[Mo['Genre'] == 'Adventure']['Budget in million'], label='Adventure',bins=15, alpha=0.5)
plt.hist(Mo[Mo['Genre'] == 'Romance']['Budget in million'], label='Romance',bins=15, alpha=0.5)

# Adding legend
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
# for more clear we stacked more
label=['Action', 'Drama', 'Thriller', 'Comedy']
plt.hist([Mo[Mo.Genre == 'Action']['Budget in million'],
          Mo[Mo.Genre == 'Drama']['Budget in million'],
          Mo[Mo.Genre == 'Thriller']['Budget in million'],
          Mo[Mo.Genre == 'Comedy']['Budget in million']],
         bins=20, stacked=True,label=label)
plt.legend(loc='upper right')
plt.show()
```

```python
#joint plot is a type of plot that combines multiple visualizations to show the relationship between two variables. 
#It typically displays a scatter plot of the two variables 
#along with additional information such as marginal distributions or a regression line.
X=sns.jointplot( data = Mo, x = 'Critics ratings', y = 'Audience ratings')
X
```

```python
X1=sns.jointplot( data = Mo, x = 'Critics ratings', y = 'Audience ratings',kind="hex")
X1
```

```python
X2=sns.jointplot( data = Mo, x = 'Critics ratings', y = 'Audience ratings',kind="reg")
X2
```

```python
X3 = sns.jointplot(data=Mo, x='Critics ratings', y='Audience ratings', hue='Genre')
X3
```

```python
##lmpot-Linear regression plot
#visualize the relationship between two variables and fit a linear regression model to the data.
Y=sns.lmplot(data=Mo, x='Critics ratings', y='Audience ratings', hue = 'Genre')
Y
```

```python
#without regession line
Y1=sns.lmplot(data=Mo, x='Critics ratings', y='Audience ratings',fit_reg=False,hue = 'Genre')
Y1
```

```python
#see in big size 12 and 1 to 1
Y2=sns.lmplot(data=Mo, x='Critics ratings', y='Audience ratings',fit_reg=False,hue = 'Genre', height=12, aspect=1)
Y2
```

```python
##KERNEL DENSITY ESTIMATE PLOT
#way to visualize the distribution of a continuous variable. 
#It represents the estimated probability density function of the variable. 
# Create a KDE plot
sns.kdeplot(data=Mo, x='Critics ratings', y='Audience ratings', shade=True, cmap='Reds')

# Display the plot
plt.show()
```
